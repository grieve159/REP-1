```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
# Upload your dataset (CSV file)
uploaded = files.upload()
# Replace 'your_file.csv' with the uploaded filename
df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

## Preprocess Data
Combine Year and Month into a single datetime column.

```python
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str))
df = df.sort_values('Date')
```

## 1. Line Chart — Rainfall Over Time

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Precipitation'], color='blue')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.title('Rainfall Over Time')
plt.show()
```

```python
# Maximum rainfall
max_rain = df.loc[df['Precipitation'].idxmax()]
max_rain[['Year','Month','Precipitation']]
```

## 2. Line Chart — Temperature vs Rainfall

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Precipitation'], 'b-', label='Rainfall')
ax2.set_ylabel('Rainfall (mm)', color='b')

plt.title('Temperature vs Rainfall')
fig.tight_layout()
plt.show()
```

**Answer:**
- Rainfall pattern with temperature: __: The blue line (rainfall) spikes periodically while the red line (temperature) shows seasonal cycles, indicating rainfall peaks coincide with higher temperatures in certain ________
- Seasonal patterns observed:  The graph shows recurring annual (seasonal) variations where temperature fluctuates smoothly and rainfall has sharp pulses, suggesting a wet season linked to temperature highs.__________

## 3. Bar Chart — Average Monthly Rainfall

```python
monthly_avg = df.groupby('Month')['Precipitation'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Precipitation', data=monthly_avg, palette='Blues')
plt.title('Average Monthly Rainfall')
plt.ylabel('Average Rainfall (mm)')
plt.show()
```

 **Answer:**
- Wettest month: _month 7______
- Driest month: _month 2_____

## 4. Scatter Plot — Rainfall vs Temperature

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Precipitation', y='Temperature', data=df)
plt.title('Rainfall vs Temperature')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Temperature (°C)')
plt.show()
```
