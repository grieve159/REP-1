Rainfall Probability
Purpose: simulate whether it rains each day.

```python
import random

# Parameters
days = 7
rain_probability = 0.3

print("Rainfall Simulation for a Week:")

for day in range(1, days + 1):
    if random.random() < rain_probability:
        print(f"Day {day}: Rain")
    else:
        print(f"Day {day}: No Rain")
```

Simulation 2: Dice Game
Purpose: see how often a 6 appears when rolling a dice.

```python
import random

rolls = 20
count_six = 0

print("Dice Roll Simulation:")

for i in range(rolls):
    dice = random.randint(1,6)
    print("Roll", i+1, ":", dice)

    if dice == 6:
        count_six += 1

print("Number of times 6 appeared:", count_six)
```

When the rolls were changed to 100 ,six appeared 21 times

Simulation 3: Population Growth
Purpose: simulate how plant population grows.

```python
population = 100
growth_rate = 0.1
years = 10

print("Population Growth Simulation")

for year in range(1, years + 1):
    population = population * (1 + growth_rate)
    print("Year", year, "Population:", int(population))
```
